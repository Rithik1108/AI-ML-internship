```python
#Dropping unnecesarry columns
zdf.drop(columns=['url','address','phone','reviews_list','menu_item','listed_in(city)','dish_liked'],inplace=True)
```

```python
#Removing duplicates
zdf.drop_duplicates(inplace=True)
zdf.duplicated().sum()
```

Cleaning rate column


```python
zdf['rate'].unique()
```

```python
def cleanrate(rate):
  if rate=='NEW' or rate=='-':
    return np.nan
  else:
    rate=str(rate).split('/')
    return float(rate[0])

zdf['rate']=zdf['rate'].apply(cleanrate)
zdf.head()
```

Filling null values in rate with its mean


```python
zdf['rate'].isnull().sum()
```

```python
mean=zdf['rate'].mean()
zdf['rate']=zdf['rate'].fillna(mean)
```

```python
zdf['rate'].isnull().sum()
```

```python
zdf.info()
```

Dropping null values(Reason:Dropping Very few null values Doesnt effect the dataset)

```python
zdf.dropna(inplace=True)
zdf.info()
```

```python
zdf.head()
```

```python
zdf.rename(columns={'approx_cost(for two people)':'2peoplecost','listed_in(type)':'type'},inplace=True)
```

```python
zdf.head()
```

Handling cost format

```python
zdf['2peoplecost'].unique()
```

```python
def handle2peoplecost(cost):
  cost=str(cost)
  if ',' in cost:
    cost=cost.replace(',','')
    return float(cost)
  else:
    return float(cost)

zdf['2peoplecost']=zdf['2peoplecost'].apply(handle2peoplecost)
zdf.head()
```

Handlling rest type(Reason:Many individual values)
[Same with other columns]

```python
zdf['rest_type'].value_counts()
```

```python
rest_type_count=zdf['rest_type'].value_counts()
```

```python
rest_type_less_than_1000=rest_type_count[rest_type_count<1000]
```

```python
def handle_rest_type(type):
  if type in rest_type_less_than_1000:
    return 'others'
  else:
    return type

zdf['rest_type']=zdf['rest_type'].apply(handle_rest_type)
```

```python
zdf['rest_type'].value_counts()
```

```python
zdf['location'].value_counts()
```

```python
location_count=zdf['location'].value_counts()
location_less_than_200=location_count[location_count<200]
```

```python
def handle_location(location):
  if location in location_less_than_200:
    return 'others'
  else:
    return location

zdf['location']=zdf['location'].apply(handle_location)
```

```python
zdf.head()
```

```python
zdf['cuisines'].value_counts()
```

```python
cuisines_count=zdf['cuisines'].value_counts()
cuisines_less_than_100=cuisines_count[cuisines_count<100]
```

```python
def handle_cuisines(cuisines):
  if cuisines in cuisines_less_than_100:
    return 'others'
  else:
    return cuisines

zdf['cuisines']=zdf['cuisines'].apply(handle_cuisines)
```
