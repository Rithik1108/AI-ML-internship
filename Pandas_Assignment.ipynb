```python
import pandas as pd
import numpy as np
```

1.  **Data Loading and Initial Inspection:**

```python
sales=pd.read_csv('/content/Sales Data.csv')
#Loading the data into the data frame
salesdf=pd.DataFrame(sales)
#.head() used to print the first 5 rows of the data frame
salesdf.head()
```

```python
#.shape is used to print the number of rows and columns of the data frame
sales.shape
```

```python
#Displayes data types of each column
salesdf.dtypes
```

2. **Data Cleaning:**


```python
#Tried to drop the rows with missing values but there arent any missing values
sdf=salesdf.dropna(inplace=True)
```

```python
sdf=salesdf.drop_duplicates()
```

```python
sdf.duplicated().sum()
```

3. **Basic Data Exploration:**

```python
#sdf.count()
sdf['Order ID'].unique().size
```

```python
#most sold item
sdf.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).idxmax()
```

```python
df=pd.to_datetime(sdf['Order Date'])
```

```python
max_date = df.max()
min_date = df.min()
print(max_date - min_date)
```

4. **Data Filtering and Selection:**

```python
#new data frame
#Create a new DataFrame containing only the sales transactions that occurred in the state of 'CA'
sdf_ca=sdf[sdf['Purchase Address'].str.split(',').str[2].str.strip(' ').str.split(' ').str[0]=='CA']
```

```python
sdf_ca
```

```python
#Filter the DataFrame to show orders where the 'Quantity Ordered' is greater than 2 and the 'Price Each' is greater than $50.
sdf[(sdf['Quantity Ordered']>2) & (sdf['Price Each']>50)]
```

```python
#Select and display the 'Order ID', 'Product', and calculated 'Total Price' for all orders.
sdf_1=sdf.rename(columns={'Sales':'Total Price'})
sdf_1[['Order ID','Product','Total Price']]
```

5. **Data Aggregation:**

```python
revenue=sdf_1['Total Price'].sum()
print('Revenue:',revenue)
```

```python
#Determine the total quantity sold for each product. Display the result sorted by quantity in descending order.
sdf_1.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)
```

```python
#Average price per product
sdf_1.groupby('Product')['Total Price'].mean()
```

6.  **Feature Engineering:**

```python
#Create a new column named 'Month' by extracting the month from the 'Order Date' column.
sdf_1['Order Date']=pd.to_datetime(sdf_1['Order Date'])
sdf_1['Month Name']=sdf_1['Order Date'].dt.month_name()
sdf_1.head()
```

```python
#Identify the top 3 months with the highest sales revenue.
sdf_1.groupby('Month Name')['Total Price'].sum().sort_values(ascending=False).head(3)
```

```python
sdf_1=sdf_1.drop(columns=['City'])
```

```python
sdf_1['City']=sdf_1['Purchase Address'].str.split(',').str[1]
```

```python
sdf_1
```

```python
#Determine which city generated the highest total revenue
sdf_1.groupby('City')['Total Price'].sum().sort_values(ascending=False).idxmax()
```
